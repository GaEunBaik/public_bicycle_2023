```python
import pandas as pd
import os

# 대여 이력 파일 경로 리스트
file_paths = [
    r'D:\따릉이\공공자전거23년\서울특별시 공공자전거 대여이력 정보_2301.csv',
    r'D:\따릉이\공공자전거23년\서울특별시 공공자전거 대여이력 정보_2302.csv',
    r'D:\따릉이\공공자전거23년\서울특별시 공공자전거 대여이력 정보_2303.csv',
    r'D:\따릉이\공공자전거23년\서울특별시 공공자전거 대여이력 정보_2304.csv',
    r'D:\따릉이\공공자전거23년\서울특별시 공공자전거 대여이력 정보_2305.csv',
    r'D:\따릉이\공공자전거23년\서울특별시 공공자전거 대여이력 정보_2306.csv',
    r'D:\따릉이\공공자전거23년\서울특별시 공공자전거 대여이력 정보_2307.csv',
    r'D:\따릉이\공공자전거23년\서울특별시 공공자전거 대여이력 정보_2308.csv',
    r'D:\따릉이\공공자전거23년\서울특별시 공공자전거 대여이력 정보_2309.csv',
    r'D:\따릉이\공공자전거23년\서울특별시 공공자전거 대여이력 정보_2310.csv',
    r'D:\따릉이\공공자전거23년\서울특별시 공공자전거 대여이력 정보_2311.csv',
    r'D:\따릉이\공공자전거23년\서울특별시 공공자전거 대여이력 정보_2312.csv'
]

# 유효한 생년 범위
valid_years = list(range(1950, 2011))

for file_path in file_paths:
    try:
        # CSV 파일 읽기 (EUC-KR로 시도)
        df = pd.read_csv(file_path, sep=',', encoding='EUC-KR', quoting=1)
    except UnicodeDecodeError:
        try:
            # CSV 파일 읽기 (CP949로 시도)
            df = pd.read_csv(file_path, sep=',', encoding='CP949', quoting=1)
        except Exception as e:
            print(f"파일 읽기 오류: {file_path}, 오류: {e}")
            continue

    # '성별' 열 처리
    df['성별'] = df['성별'].replace({r'\N': None, 'f': 'F', 'F': 'F', 'm': 'M', 'M': 'M'})
    df = df.dropna(subset=['성별'])

    # '생년' 열 처리
    df['생년'] = df['생년'].replace({r'\N': None, '0': None})
    df = df.dropna(subset=['생년'])
    df['생년'] = df['생년'].astype(int)
    df = df[df['생년'].isin(valid_years)]

    # '대여일시' 열에서 2323년 데이터 삭제
    df = df[~df['대여일시'].str.contains('2323', na=False)]

    # '자전거구분', '반납거치대', '반납대여소번호', '반납거치대' 열의 \N 및 NaN 삭제
    columns_to_clean = ['자전거구분', '반납거치대', '반납대여소번호', '반납거치대']
    for col in columns_to_clean:
        if col in df.columns:
            df[col] = df[col].replace(r'\N', None)
            df = df.dropna(subset=[col])

    # '이용시간(분)'과 '이용거리(M)' 값이 1 미만인 행 삭제
    df = df[(df['이용시간(분)'] >= 1) & (df['이용거리(M)'] >= 1)]

    # 일련번호 열 추가
    df.insert(0, '일련번호', range(1, len(df) + 1))

    # 처리된 데이터를 새로운 파일로 저장
    output_path = os.path.join(r'D:\따릉이\공공자전거23년\processed', os.path.basename(file_path))
    os.makedirs(os.path.dirname(output_path), exist_ok=True)
    df.to_csv(output_path, index=False, encoding='utf-8-sig')

    print(f"{file_path} 파일 처리 완료 및 저장.")
```
